```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.formula.api import ols
import plotly.express as px
```

### 1. Carga de datos

```python
df = pd.read_csv('/content/drive/MyDrive/Teaching/UPSA/2025.01/MOD/forecasting/sales.csv')
df.info()
```

### 2. Preprocesado

```python
# Period a formato fecha
df['period'] = pd.to_datetime(df['period'], format='%m/%d/%Y')
df.info()
```

```python
# df de entrenamiento
df_training = df[df['period'] <= '2015-12-01'].copy()
```

```python
# DescomposiciÃ³n de la serie
decomposed = seasonal_decompose(
    df_training.set_index('period')['sales'],
    model='additive',
    period=12
)
```

```python
decomposed.plot();
```

### 3. ConstrucciÃ³n del modelo

```python
df_model = df_training.copy()
df_model['trend'] = range(1, len(df_model) + 1)
```

```python
# Modelo de regresiÃ³n
model = ols(
    formula = 'sales ~ trend',
    data = df_model
).fit()
```

```python
print(model.summary())
```

### 4. DiagnÃ³stico del modelo

```python
df_model['predicted'] = model.predict()
df_model.head()
```

```python
px.scatter(
    df_model,
    x = 'sales',
    y = 'predicted',
    title = 'Sales vs Predicted'
).add_shape(
    type = 'line',
    x0 = df_model['sales'].min(),
    y0 = df_model['sales'].min(),
    x1 = df_model['sales'].max(),
    y1 = df_model['sales'].max()
)
```

```python
# Serie temporal + predicciones
px.line(
    df_model,
    x = 'period',
    y = ['sales', 'predicted'],
    title = 'Sales vs Predicted'
)
```

### 5. ValidaciÃ³n de supuestos

```python
residuals = model.resid
residuals.head()
```
